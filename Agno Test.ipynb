```python
# Import the Ollama library to interact with the locally installed DeepSeek R1 model
import ollama

# Define the model name (DeepSeek R1 1.5B as installed via Ollama)
model_name = "deepseek-r1:1.5b"

# Define a sample question or prompt to send to the model
prompt = "What is the capital of France, and why is it famous?"

# Use the Ollama chat function to get a response from the model
response = ollama.chat(
    model=model_name,
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ]
)

# Extract and print the model's response
answer = response["message"]["content"]
print("DeepSeek R1 Response:")
print(answer)
```
